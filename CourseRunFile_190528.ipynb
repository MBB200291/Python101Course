```python
# exercise
## 在沒有sub-folder的檔案中，找出所有txt檔案，將前二十行寫至同一個檔案，並且tab分隔的第一欄寫上檔案名稱
```

```python
path_dir = './python101_demo_data/data_wo_sudir/'
list_target_files = [x for x in os.listdir(path_dir) if x.endswith('.txt')]

list_contents = []  # [(filename, top20lines), ()]
for f in list_target_files:
    list_content20 = []
    with open(os.path.join(path_dir,f),'r',encoding='utf8') as inputfile:
        for i in range(20):
            list_content20.append((f, inputfile.readline().strip('\n')))
    list_contents += list_content20

with open('result_top20.txt', 'w', encoding='utf8') as ouputf:
    for f,c in list_contents:
        ouputf.write(f+'\t'+c+'\n')
```

```python
# method2

with open('./result_top20_2.txt', 'w', encoding='utf8') as outputf:
    for f in [x for x in os.listdir(path_dir) if x.endswith('.txt')]:
        print(f)
        with open(os.path.join(path_dir,f),'r',encoding='utf8') as inputfile:
            for i in range(20):
                line = inputfile.readline().strip('\n')
                outputf.write(f'{f}\t{line}\n')
        outputf.write('*'*100+'\n')
```

```python
dict_index = {1:2, 2:1, 3:-1, 4:0}     # index: 1=>(+), 2=>(=), 3=>(-), 4=>(0) ; value_means: 2=>positive, 1=>neutral, 2=> negative, 0=>other
def ComputeDirection(list_seq, dict_index=dict_index):
    result = 1
    for i in list_seq:
        result *= dict_index[i]
    print(result)
    if result < 0:
        return 'negative'
    elif result == 0:
        return 'other'
    elif result == 1:
        return 'neutral'
    elif result > 1:
        return 'positive'
```

```python
ComputeDirection([1,2,1,3])
```

```python
# excercise 2 ==> 只寫出有"amplicon"的該行

with open('./result_ampliconline.txt', 'w', encoding='utf8') as outputf:
    for f in [x for x in os.listdir(path_dir) if x.endswith('.txt')]:
        print(f)
        with open(os.path.join(path_dir,f),'r',encoding='utf8') as inputfile:
            count = 0
            for i in [x for x in inputfile.read().splitlines() if 'amplicon' in x]:
                count += 1
                outputf.write(f'{f}\t{i}\n')
            print('> number of lines:', count)
        outputf.write('*'*100+'\n')
```

```python
def funA(a, fun):
    return fun(a+1)
def X(b, c):
    return b+c
```

```python
# method1: 
funA(1, lambda x:X(x, 2))
```

```python
# method2: 
def X_p(a):
    return X(a, 2)
funA(1, X_p)
```
