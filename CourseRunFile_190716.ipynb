```python
import os
import re
```

```python
def ReadFileAsList(Path):
    with open(Path, 'r', encoding='utf8', newline='') as inp:
        return inp.read().splitlines()
```

```python
def GetREcontent(re_match, GrouP=1):
    if re_match:
        return re_match.group(GrouP)
    else:
        return ''
```

```python
pat_ID = re.compile(r'\t(chr[\w>]+)\t')
pat_VAF = re.compile(r'\tVAF=([01]\.\d+);')
pat_DP = re.compile(r';DP=(\d+);')
def ParseCol(line):
    def GetREcontent(re_match, GrouP=1):
        if re_match:
            return re_match.group(GrouP)
        else:
            return ''
    ID = GetREcontent(pat_ID.search(line), 1)
    VAF = GetREcontent(pat_VAF.search(line), 1)
    DP = GetREcontent(pat_DP.search(line), 1)
    return ID,VAF,DP
```

```python
def getVariantID_VAF_DP(file):
    dict_id__VAF_DP = {}
    for l in ReadFileAsList(file):
        ID, VAF, DP = ParseCol(l)
        if ID != '':
            dict_id__VAF_DP[ID] = (VAF, DP)
        #print(i, ParseCol(l))
    return dict_id__VAF_DP
```

```python
li_target_files = [x for x in os.listdir() if x.startswith('snp_testing_')]
dict_result = {}
for i in li_target_files:
    dict_result[i] = getVariantID_VAF_DP(i)
```

```python
# output
for f in dict_result:
    path_out = os.path.splitext(f)[0]+'_ID_VAF_DP.txt'
    with open(path_out, 'w', encoding='utf8', newline='') as oup:
        print(f)
        print(path_out)
        for ID, vaf_dp in dict_result[f].items():
            oup.write(f'{f}\t{ID}\t{vaf_dp[0]}\t{vaf_dp[1]}\n')    
```
